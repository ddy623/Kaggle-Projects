Analyzing Data 

```python
# Identify columns with numeric data you want to aggregate
numeric_columns = ['Age (Years)', 'Weight (kg)']
```

```python
# Calculate mean for numeric columns only
df[numeric_columns].mean()
```

```python
df['Breed'].value_counts()
```

```python
df.groupby('Color').agg({'Age (Years)':['mean','max', 'count','sum'] })
```

```python
df.groupby('Gender').agg({'Age (Years)':['mean','max', 'count','sum'] })
```

```python
df.groupby('Breed').agg({'Age (Years)':['mean','max', 'count','sum'] })
```

The data shows that there are more female than male dogs, as there was 1520 female and 1480 males. The average age of 7 to 8 years with the max age of 14 years old.  It also showed many breeds and many different colors.  The majority of the dogs in the sample were Rottweilers, as the sample had 118 in the sample.  

Explore Data Analysis


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Create a countplot for categorical data
# Seaborn countplot
fig, ax = plt.subplots(figsize=(12, 6))
sns.countplot(data=df, x='Breed', ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

```python
#Create a countplot for categorical data
# Seaborn countplot
fig, ax = plt.subplots(figsize=(12, 6))
sns.countplot(data=df, x='Gender', ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```
